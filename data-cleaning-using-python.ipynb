```python
import pandas as pd
```

```python
Data = pd.read_excel('/kaggle/input/cleaning-the-data/Customer Call List temp.xlsx')
Data
```

```python
Data.fillna(' ', inplace=True)
```

#  1- Remove duplicates

```python
Data = Data.drop_duplicates()
Data
```

# 2- drop unusefull columns 

```python
#2- drop not usefull columns 
Data = Data.drop(columns = 'Not_Useful_Column')
Data
```

# Standarize Last_Name column

```python
# Standarize Last_Name column
Data['Last_Name'] = Data['Last_Name'].str.strip("./|_")
Data
```

```python
#remonve any chahracter except nubers
Data['Phone_Number'] = Data['Phone_Number'].replace('[^0-9]', '', regex=True).fillna('')
Data
```

# look on Phone_Number column

```python
#look on Phone_Number column
#convert it into strig
Data['Phone_Number'] = Data['Phone_Number'].apply(lambda x : str(x))
```

```python
Data['Phone_Number'].apply(lambda x : x[0:3] + '-' + x[3:6] + '-' + x[6:])
```

```python
Data['Phone_Number'] = Data['Phone_Number'].apply(lambda x : x[0:3] + '-' + x[3:6] + '-' + x[6:])
Data
```

```python
Data['Phone_Number'] = Data['Phone_Number'].str.replace('--', '')
Data
```

# split Address column 

```python
Data[["Street", "State", "Zip_Code"]] = Data['Address'].str.split(',', expand=True)
Data
```

# Replace any Y to Yes and any N to No

```python
Data["Paying Customer"] = Data["Paying Customer"].replace({"Y": "Yes", "N": "No"})
Data["Do_Not_Contact"] = Data["Do_Not_Contact"].replace({"Y": "Yes", "N": "No"})
Data
```

```python
Data.fillna(' ', inplace=True)
Data
```

# Drop unusfull rows

```python
#drop rows that Do_Not_Contact== Yes or blank
for x in Data.index:
    if Data.loc[x, "Do_Not_Contact"] == 'Yes' or  Data.loc[x, "Do_Not_Contact"] == ' ':
        Data.drop(x, inplace = True)

Data
```
